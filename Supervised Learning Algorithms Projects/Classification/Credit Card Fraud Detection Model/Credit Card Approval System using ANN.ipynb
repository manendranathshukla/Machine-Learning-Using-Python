##  Credit Card Fraud Detection System
### is a problem statement given by Bank sector to check the given applicant can pay the money after getting the Credit Card or its a Fraud?


```python
# importing libraries
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#importing dataset
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
X=df.iloc[:,0:29].values
y=df.iloc[:,29:].values
```

```python
X.shape
```

```python
y.shape
```

```python
y
```

```python
#Spliting the dataset into train and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Initilizing the ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
#Adding first and hidden layer 
model.add(Dense(15,kernel_initializer='uniform',activation='relu',input_dim=29))
#adding 2nd hidden layer
model.add(Dense(15,kernel_initializer='uniform',activation='relu'))
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['Accuracy'])
```

```python
model.summary()
```

```python
history=model.fit(x_train,y_train,batch_size=100,epochs=100,validation_data=(x_test,y_test))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Model Performance")
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
