## Google Stock Price Prediction using Recurrent Neural Network(RNN)
This is a type of regression problem where we have to predict the stock price by training the dataset generated by Google Finance.
Happy Learning!!

```python
# Part 1 - Data Preprocessing
#importing the libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset 
traing_set=pd.read_csv("Google_Stock_Price_Train.csv")
test_set = pd.read_csv("Google_Stock_Price_Test.csv")
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
traing_set.head()
```

```python
traing_set.shape
```

```python
traing_set = traing_set.iloc[:,1:2].values
```

```python
traing_set
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
sc=MinMaxScaler()
traing_set = sc.fit_transform(traing_set)
```

```python
traing_set
```

```python
#Getting the inputs and the outputs 
x_train = traing_set[0:1257]
y_train = traing_set[1:1258]
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# Reshaping the inputs
x_train = np.reshape(x_train,(1257,1,1))
```

```python
#Part 2 - Building the RNN
#importing the keras libraries and packages
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
#Initiliazing the RNN
regressor = Sequential()
```

```python
#Adding the input layer and the LSTM layer
regressor.add(LSTM(units=4,activation='sigmoid',input_shape = (None,1)))
```

```python
#Adding the output layer 
regressor.add(Dense(units=1))
```

```python
#Compiling the RNN
regressor.compile(optimizer='adam',loss='mean_squared_error')
```

```python
#Fitting the  RNN to the Training Set
regressor.fit(x_train,y_train,batch_size=32,epochs=200)
```

```python
#Part 3- Making the predictions and visualizing the results
#Getting the real stock price of 2017
```

```python
real_stock_price = test_set.iloc[:,1:2].values
```

```python
real_stock_price
```

```python
#Getting the predicted stocks price of 2017
inputs = real_stock_price
inputs=sc.fit_transform(inputs)
inputs = np.reshape(inputs,(20,1,1))
```

```python
predicted_stock_price = regressor.predict(inputs)
```

```python
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
predicted_stock_price
```

```python
#Visualizing the results
plt.plot(real_stock_price,color="red",label='Real Google Stock Price')
plt.plot(predicted_stock_price,color="blue",label='Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```

```python
#Getting the real stock price of 2012 -2016
real_stock_price_train = pd.read_csv("Google_Stock_Price_Train.csv")
real_stock_price_train = real_stock_price_train.iloc[:,1:2].values
```

```python
inputs1 = real_stock_price_train
inputs1=sc.fit_transform(inputs1)
inputs1 = np.reshape(inputs1,(1258,1,1))
```

```python
#Getting the predicted stock of 2012 - 2016
predicted_stock_price_train = regressor.predict(inputs1)
predicted_stock_price_train=sc.inverse_transform(predicted_stock_price_train)
```

```python
#Visualizing the results 
predicted_stock_price_train
```

```python
#Visualizing the results
plt.plot(real_stock_price_train,color="red",label='Real Google Stock Price')
plt.plot(predicted_stock_price_train,color="blue",label='Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```

```python
# Part 4 - Evaluating the RNN
import math
from sklearn.metrics import mean_squared_error
rmse= math.sqrt(mean_squared_error(real_stock_price_train,predicted_stock_price_train))
```
